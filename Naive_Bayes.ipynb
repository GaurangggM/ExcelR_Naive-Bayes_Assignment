# `Splitting the data`

```python
from sklearn.model_selection import train_test_split
```

```python
x_train = train.iloc[:,0:13]
y_train = train.iloc[:,-1] 
x_test = test.iloc[:,0:13]
y_test = test.iloc[:,-1]
```

# `Applying Naive Bayes`

```python
from sklearn.naive_bayes import MultinomialNB as MB
from sklearn.naive_bayes import CategoricalNB as CNB
from sklearn.naive_bayes import GaussianNB as GB
```

1) Multinomial 

```python
mb=MB()
mb.fit(x_train,y_train)
```

```python
train_pred_mb=mb.predict(x_train)
accuracy_train_mb=np.mean(train_pred_mb==y_train)
```

```python
test_pred_mb=mb.predict(x_test)
accuracy_test_mb=np.mean(test_pred_mb==y_test)
```

```python
print('Accuracy on train data set = ',accuracy_train_mb)
print('Accuracy on test data set = ',accuracy_test_mb)
```

2) Categorical

```python
cnb=CNB()
cnb.fit(x_train,y_train)
```

```python
train_pred_cnb=cnb.predict(x_train)
accuracy_train_cnb=np.mean(train_pred_cnb==y_train)
```

```python
test_pred_cnb=cnb.predict(x_test)
accuracy_test_cnb=np.mean(test_pred_cnb==y_test)
```

```python
print('Accuracy on train data set = ',accuracy_train_cnb)
print('Accuracy on test data set = ',accuracy_test_cnb)
```

3) Gaussian

```python
gb=GB()
gb.fit(x_train,y_train)
```

```python
train_pred_gb=gb.predict(x_train)
accuracy_train_gb=np.mean(train_pred_gb==y_train)
```

```python
test_pred_gb=gb.predict(x_test)
accuracy_test_gb=np.mean(test_pred_gb==y_test)
```

```python
print('Accuracy on train data set = ',accuracy_train_gb)
print('Accuracy on test data set = ',accuracy_test_gb)
```

# `Categorical Naive Bayes Gives us better accuracy i.e 89.27% on Trainind Dataset and 85.05 on Test Dataset`
